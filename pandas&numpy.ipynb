```python
#Q. give a positive no 'n' greater than 2 create a numpy array of size(n*n) with all zeroes and ones such that ones make a sahape

import numpy as np

def create_shape_array(n):
    arr = np.zeros((n, n), dtype=int)
    mid = n // 2

    arr[mid, :] = 1
    arr[:, mid] = 1

    return arr

n = 5
result = create_shape_array(n)
print(result)


arr=np.zeros((n,n),dtype='int')



```

```python
import pandas as pd

# DataFrames
df1 = pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C4','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})

df2 = pd.DataFrame({
    'E':['E4','E5','E6','E7'],
    'F':['F4','F5','F6','F7'],
    'C':['C4','C1','C2','C3'],
    'G':['G4','G5','G6','G7']
})

# Merge
merge_df = pd.merge(df1, df2, on='C', how='inner')

print(merge_df)
```
