```python
#file Handling
#create a text file
with open ("sample.txt","w") as file:
    file.write("Hello Students\n")
    file.write("Welcome to DA course\n")
    file.write("Bhagya is a trainee\n")

```

```python
#read a file
file=open("sample.txt","r")
```

```python
for line in file:
  print(line)
```

```python
file=open("sample.txt","r")
print(file.read())
```

```python
file=open("sample.txt","r")
print(file.read(11))
```

```python
file=open("sample.txt","r")
print(file.readline())
```

```python
#writing a file
with open ("sample1.txt","w") as file:
    file.write("Hello  dear Students\n")
    file.write("Welcome to DA course data science\n")
    file.write("Bhagya is a trainee\n")
    file.close()
```

```python
file=open("sample1.txt","r")

print(file.read())
```

```python
#append operation
with open ("sample1.txt","a") as file:
    file.write("Hello  dear Students\n")
    file.write("Welcome to DA course data science\n")
    file.write("Bhagya is a trainee\n")
```

```python
file=open("sample1.txt","r")
```

```python
for line in file:
  print(line)
```
