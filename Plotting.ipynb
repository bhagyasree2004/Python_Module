```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df.plot(x='year',y='Exchange',kind='barh')
```

```python
df.plot(x='year',y='Exchange',kind='pie')
```

```python
#another method for using matplotlib

plt.scatter(df['year'],df['Exchange'])
plt.xlabel('year')
plt.ylabel('Exchange')
plt.show()
```

```python
df=pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
plt.bar(df['gender'],df['MonthlyCharges'].max(),color='blue')

plt.title('Monthly Charges')
plt.xlabel('Gender')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
plt.bar(df['gender'],df['MonthlyCharges'].max(),color='cyan',edgecolor='Orange')

plt.title('Monthly Charges',fontdict={"fontsize":20,"color":"Green"})
plt.xlabel('Gender')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
plt.scatter(df['MonthlyCharges'],df['tenure'])
plt.title('Monthly Charges vs Tenure')
plt.xlabel('Monthly Charges')
plt.ylabel('Tenure')
plt.show()
```

```python
#histogram
plt.hist(df['tenure'],bins=30)
plt.title('Tenure Distribution')
plt.xlabel('Tenure')
plt.ylabel('Frequency')
plt.show()
```

```python
#pie chart
labels=['javascript','python','R','java']
sizes=[15,10,45,60]
explode=[0,0.2,0,0]


fig1, ax1 = plt.subplots()




ax1.pie(sizes, labels=labels, explode=explode, shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
#plot using seaborn
import seaborn as sns


sns.countplot(x="Churn", data=df)
plt.title("Customer Churn Count")
plt.show()



```

```python
sns.countplot(x="gender", data=df)
plt.title("Customer Gender Count")
plt.show()
```

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
sns.histplot(df["tenure"], bins=30)
plt.title("Tenure Distribution")
plt.show()
```

```python
sns.set()

# Count churn values
data = df["Churn"].value_counts()

# Pie chart
plt.pie(data, labels=data.index, autopct='%1.1f%%')
plt.title("Churn Distribution")

plt.show()
```

```python
#scatter plot
sns.scatterplot(x="MonthlyCharges", y="tenure", data=df)
plt.title("Monthly Charges vs Tenure")
plt.show()
```

```python
#LINE CHART
sns.lineplot(x="tenure", y="MonthlyCharges",data=df)
plt.title("Monthly Charges vs Tenure")
plt.show()
```
