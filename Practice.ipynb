```python
def string_palindrome(s):
    if s == s[::-1]:
        print("palindrome")
    else:
        print("NOt palindrome")
string_palindrome("madam")
```

```python
def number_palindrome(n):
    original = n
    reverse = 0

    while n > 0:
        digit = n % 10
        reverse = reverse * 10 + digit
        n = n // 10

    if original == reverse:
        print("palindrome")
    else:
        print("NOt palindrome")

number_palindrome(121)
```

```python
def reverse(n):
    reverse = 0

    while n > 0:
        digit = n % 10
        reverse = reverse * 10 + digit
        n = n // 10

    return reverse
reverse(12345)
```

```python
def reverse_num(n):
    return int(str(n)[::-1])

print(reverse_num(5678))
```
